```python
%matplotlib inline
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

img=Image.open('lena_gray_256.tif')
plt.subplot(3,2,1)
img
```

```python
numpydata= np.asarray(img)
print(numpydata)
```

```python
plt.subplot(3,2,2)
a=plt.hist(numpydata)
a
```

```python
print("After adding noise:")
noise=np.random.normal(0,1,(numpydata.shape))
noiseImg=Image.fromarray(noise,'L')

plt.subplot(3,2,3)
noiseImg
```

```python
new_image=numpydata + noise
print(new_image)
```

```python

plt.subplot(3,2,4)
plt.hist(noiseImg)
```

```python
pic=Image.fromarray(new_image,'L')

plt.subplot(3,2,5)
pic
```

```python

plt.subplot(3,2,6)
plt.hist(new_image)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(3,2,1)
ax1.imshow(img)
ax2 = fig.add_subplot(3,2,2)
plt.hist(numpydata)
ax3 = fig.add_subplot(3,2,3)
ax3.imshow(noiseImg)
ax4 = fig.add_subplot(3,2,4)
plt.hist(noiseImg)
ax5=fig.add_subplot(3,2,5)
ax5.imshow(pic)
ax6=fig.add_subplot(3,2,6)
plt.hist(new_image)
```
